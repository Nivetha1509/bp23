```python
def is_symmetrical(n):
    org=n
    s=0
    while(org!=0):
        rem=org%10
        s=(s*10)+rem
        org=org//10
    if(s==n):
        return True
    else:
        return False
print(is_symmetrical(7227))
print(is_symmetrical(12567))        
print(is_symmetrical(44444444))      
print(is_symmetrical(9939))
print(is_symmetrical(1112111))
```

```python
def multiply_nums(s):
    p=1
    for i in s.split(","):
        p=p*int(i)
    print(p)        
multiply_nums("2, 3")   
multiply_nums("1, 2, 3, 4")
multiply_nums("54, 75, 453, 0")
multiply_nums("10, -2")
```

```python
def square_digits(n):
    s=""
    for i in str(n):
        y=int(i)*int(i)
        s=s+str(y)
    print(s)
square_digits(9119) 
square_digits(2483)
square_digits(3212)        
        
```

```python
def setify(l):
    s=set(l)
    n=list(s)
    print(n)
setify([1, 3, 3, 5, 5])
setify([4, 4, 4, 4])
setify([5, 7, 8, 9, 10, 15])
setify([3, 3, 3, 2, 1])
```

```python
def mean(n):
    s=0
    c=0
    for i in str(n):
        s=s+int(i)
        c=c+1
    mean=s/c
    print(round(mean))
mean(42)     
mean(12345)        
mean(666)        
```
